```python
# Extraindo o ano da data / Contar a tentativa por ano
df['ano'] = df['data'].dt.year

# Contar tentativas de suicídio por ano
tentativas_por_ano = df['ano'].value_counts().sort_index()

# Mostrando os resultados
print(tentativas_por_ano)
```

```python
# INdentificar o ano com menos tentativas
ano_mais_tentativas = tentativas_por_ano.idxmax()
ano_menos_tentativas = tentativas_por_ano.idxmin()

print(f"Ano com mais tentativas: {ano_mais_tentativas} ({tentativas_por_ano.max()} tentativas)")
print(f"Ano com menos tentativas: {ano_menos_tentativas} ({tentativas_por_ano.min()} tentativas)")
```

```python
# Calcular a porcentagem de variação entre os anos
variacao_percentual = (tentativas_por_ano.diff() / tentativas_por_ano.shift(1)) * 100

# Mostrando a variação percentual
print(variacao_percentual)
```
